```python
import gradio as gr
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch
model_name = "microsoft/DialoGPT-medium"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
```

```python
def generate_response(inp)->str:
    input_ids = tokenizer.encode(inp + tokenizer.eos_token, return_tensors="pt")
    chat_history_ids = model.generate(input_ids, max_length=1000, pad_token_id=tokenizer.eos_token_id)
    response = tokenizer.decode(chat_history_ids[:, input_ids.shape[-1]:][0], skip_special_tokens=True)
    return response
```

```python
generate_response("Hello")
```

```python
# Create a Gradio interface using the updated API
interface = gr.Interface(
    fn=generate_response,
    inputs=gr.Textbox(lines=2, placeholder="Type your message here..."),
    outputs=gr.Textbox(label="Response"),
    title="Chatbot",
    description="Enter a message and the chatbot will respond."
)
```

```python
# Launch the interface
interface.launch()
```
